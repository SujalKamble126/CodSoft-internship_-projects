```python
import numpy as np # linear algebra
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('IRIS.csv')
df
```

```python
df['species'].unique()
```

```python
df.boxplot(column = 'sepal_length')
plt.show()
```

```python
df.boxplot(column = 'sepal_width')
plt.show()
```

```python
df.boxplot(column = 'petal_length')
plt.show()
```

```python
df.boxplot(column = 'petal_width')
plt.show()
```

```python
sns.kdeplot(data=df, x = 'sepal_length',hue= 'species')
plt.title("Sepal Length with Species")
plt.xlabel("Sepal Lenght")
plt.ylabel("Probability Density")
plt.show()
```

```python
sns.kdeplot(data=df, x = 'sepal_width',hue= 'species')
plt.title("Sepal Width with Species")
plt.xlabel("Sepal Widht")
plt.ylabel("Probability Density")
plt.show()
```

```python
sns.kdeplot(data=df, x = 'petal_length',hue= 'species')
plt.title("Petal Length with Species")
plt.xlabel("Petal Lenght")
plt.ylabel("Probability Density")
plt.show()
```

```python
sns.ecdfplot(data = df, x = 'petal_length', hue = 'species')
sns.kdeplot(data=df, x = 'petal_length',hue= 'species')
plt.title(" Petal Length with Species")
plt.xlabel("Petal Lenght")
plt.ylabel("Probability Density")
plt.show()
```
